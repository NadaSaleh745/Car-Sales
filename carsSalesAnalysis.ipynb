```python
from pycountry_convert import country_alpha2_to_continent_code, country_name_to_country_alpha2
def get_continent(col):
    try:
        cn_a2_code =  country_name_to_country_alpha2(col)
    except:
        cn_a2_code = 'Unknown' 
    try:
        cn_continent = country_alpha2_to_continent_code(cn_a2_code)
    except:
        cn_continent = 'Unknown' 
    return pd.Series([cn_a2_code, cn_continent])

df[['country_code', 'continent']] = df['country'].apply(get_continent)
```

```python
from geopy.geocoders import Nominatim

geolocator = Nominatim(user_agent="nadasaleh745@myserver.com", timeout=10)

def get_lat_lon(country_code):
    location = geolocator.geocode(country_code)
    if location:
        return location.latitude, location.longitude
    else:
        return None, None
```

```python
def remove_outliers(df, exclude_columns=None):
    df_filtered = df.copy()
    if exclude_columns is None:
        exclude_columns = []
    
    for col in df_filtered.select_dtypes(include=np.number).columns:
        if col in exclude_columns:
            continue
        
        q1 = df_filtered[col].quantile(0.25)
        q3 = df_filtered[col].quantile(0.75)
        iqr = q3 - q1
        lb = q1 - 1.5 * iqr
        ub = q3 + 1.5 * iqr
        df_filtered = df_filtered[(df_filtered[col] >= lb) & (df_filtered[col] <= ub)]
    
    return df_filtered

dfNoOutliers = remove_outliers(df, exclude_columns=['annual Salary'])
dfNoOutliers

```

```python
dfNoOutliers = dfNoOutliers[['age', 'annual Salary', 'net worth', 'car purchase amount']]
dfNoOutliers
```

```python
from sklearn.model_selection import train_test_split
y = dfNoOutliers['car purchase amount']
X = dfNoOutliers.drop('car purchase amount', axis=1)

X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size=0.3, random_state=123)
```

```python
print("Training feature set size:",X_train.shape)
print("Test feature set size:",X_test.shape)
print("Training label set size:",Y_train.shape)
print("Test label set size:",Y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
lm = LinearRegression()
lm.fit(X_train, Y_train)
trainPred = lm.predict(X_train)
predictions = lm.predict(X_test)
```

```python
print("R-squared for training:",(r2_score(Y_train,trainPred)))
print("Mean square error (MSE) for training:", mean_squared_error(Y_train,trainPred))

print("R-squared for testing:",(r2_score(Y_test,predictions)))
print("Mean square error (MSE) for testing:", mean_squared_error(Y_test,predictions))
```
